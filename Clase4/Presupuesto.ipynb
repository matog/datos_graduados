```python
import pandas as pd
import seaborn as sns
```

```python
# Como las columnas con los créditos presupuestarios están definidas como object, hay que convertirlas float
# Primero cambiamos el separador de decimal por '.'.
df[['credito_presupuestado', 'credito_vigente','credito_comprometido','credito_devengado','credito_pagado']] = df[['credito_presupuestado', 'credito_vigente','credito_comprometido','credito_devengado','credito_pagado']].apply(lambda x: x.str.replace(',','.'))
```

```python
# Ahora las convertimos a float
df[['credito_presupuestado', 'credito_vigente','credito_comprometido','credito_devengado','credito_pagado']] = df[['credito_presupuestado', 'credito_vigente','credito_comprometido','credito_devengado','credito_pagado']].astype(float)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(x='impacto_presupuestario_mes', 
             y='credito_devengado',
             data=df.groupby(['jurisdiccion_desc','impacto_presupuestario_mes'])['credito_devengado','credito_vigente'].sum().reset_index()
            )
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(x='impacto_presupuestario_mes', 
             y='credito_devengado',
             hue="jurisdiccion_desc",
             data=df.groupby(['jurisdiccion_desc','impacto_presupuestario_mes'])['credito_devengado','credito_vigente'].sum().reset_index()
            ).set(title='Ejecución presupuestaria mensual por SAF')

# Ubicamos la leyenda del gráfico. Si borramos esta linea queda superpuesta.

plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0);
```

```python
# Initialize a grid of plots with an Axes for each walk
df_1 = df.groupby(['jurisdiccion_desc','impacto_presupuestario_mes'])['credito_devengado','credito_vigente'].sum().reset_index()
g = sns.FacetGrid(df_1, col = 'impacto_presupuestario_mes')
g.map(plt.scatter, "impacto_presupuestario_mes", "credito_devengado")
plt.show()
```

```python
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=df_1,
    x="impacto_presupuestario_mes", y="jurisdiccion_desc",
    #hue="year", 
    size="credito_devengado",
    palette=cmap, sizes=(10, 200),
)
```

```python
df_2 = pd.melt(df_1, 
               id_vars =['jurisdiccion_desc','impacto_presupuestario_mes'], 
               var_name = 'tipo', 
               value_name = 'valor' )
df_2
```

```python
df_2.tipo.value_counts()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(x='impacto_presupuestario_mes', 
             y='valor',
             hue="tipo",
             data=df_2,
            ).set(title='Ejecución presupuestaria mensual por SAF')

# Ubicamos la leyenda del gráfico. Si borramos esta linea queda superpuesta.

plt.legend(bbox_to_anchor=(1.02, 1), 
           loc='upper left', 
           );
```
